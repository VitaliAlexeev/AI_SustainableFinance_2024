# NLP module

```python
import utsnlp
```

## Get text from Wikipedia

```python
topic = "Machine learning"
#topic = "Kullback–Leibler divergence"
```

```python
text = utsnlp.download_wikipedia_text(topic)
text[:1000]
```

```python
options = {
    'summary': True,
    'include_url': True,
    'include_sections': True,
    'include_links': True,
    'include_categories': True,
    'include_languages': True,
    'save_to_file': True,
    'file_name': 'machine_learning_info.txt'
}
text = utsnlp.download_wikipedia_text(topic, **options)
```

## Preprocessing text

```python
# Example usage
#text = "The quick brown fox jumps over the lazy dog."
#text = "We'll go. I'll go but you can't go myself."  #  contractions.fix(text)
#text = "Visit https://example.com for more info. <b>This is bold text</b> and this is a LaTeX formula: $E=mc^2$."
text = "This is a sample text with (parentheses), [brackets], {curly brackets}, $LaTeX$ expression, and <html> tags."
```

```python
cleaned_text = utsnlp.preprocess_text(text)
print('Raw text:')
print(text[:1000])
print('\nPreprocessed text:')
print(cleaned_text[:1000])
```

```python
options = {
    'remove_text_in_brackets':False,
    'remove_url':True,
    'remove_html_tags':True,
    'to_lower':False,
    'expand_contractions':True,
    'remove_punctuation':False, 
    'remove_digits':False, 
    'remove_stopwords':False, 
    'lemmatize':False, 
    'stem':False,
    'custom_stopwords':None,
    'remove_text_in_brackets':False,
    'custom_brackets':'''{}()[]'''
}

cleaned_text = utsnlp.preprocess_text(text, **options)
print('Raw text:')
print(text[:1000])
print('\nPreprocessed text:')
print(cleaned_text[:1000])
```
